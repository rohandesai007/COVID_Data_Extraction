```python
import pandas as pd
import requests

url = "https://www.worldometers.info/coronavirus/country/us/"
r = requests.get(url)
df_list = pd.read_html(r.text) # this parses all the tables in webpages to a list
df = df_list[0]
df.head()
```

```python
del df['Source']
del df['Projections']
```
